# Zadanie 3

```python
import math
def solve(a,b,y):
    return round(0.5*int(a)*int(b)*math.sin(math.radians(int(y))),3)
f = open("wynik.txt", "w")
with open("dane.txt") as f2:
    for line in f2:
        a,b,y = line.split()
        p = solve(a,b,y)
        f.write(f"{a},{b},{y},{p}\n")
f.close()
```

# Zadanie 4

```python
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

class Cart:
    def __init__(self):
        self.shoppinglist = [Product("marchew", 2, 10), Product("ziemniaki", 1, 12), Product("jogurt", 5, 1)]
    def add(self, name):
        for product in self.shoppinglist:
            if product.name == name:
                product.quantity += 1

    def remove(self, name):
        for product in self.shoppinglist:
            if product.name == name:
                product.quantity -= 1
                if product.quantity < 0:
                    product.quantity = 0

    def total_price(self):
        price = 0
        for product in self.shoppinglist:
            price += product.price*product.quantity
        return price

    def print_cart(self):
        for product in self.shoppinglist:
            print(product.name, product.price, product.quantity)

koszyk = Cart()
print("Poczatkowa zawartosc koszyka:")
koszyk.print_cart()
print("Poczatkowa przeliczona wartosc koszyka:")
print(koszyk.total_price())
print("\n")
koszyk.add("marchew")
koszyk.remove("ziemniaki")
koszyk.add("jogurt")
koszyk.add("banan")
print("\n")
print("Zawartosc koszyka po wywolaniu metod:")
koszyk.print_cart()
print("Wartosc koszyka po wywolaniu metod:")
print(koszyk.total_price())


```

# Zadanie 5

```python
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def __str__(self):
        return f"Produkt: {self.name}, cena: {self.price} zl "

class Cart:
    def __init__(self):
        self.shoppinglist = [Product("marchew", 2, 10), Product("ziemniaki", 1, 12), Product("jogurt", 5, 1)]
    def __str__(self):
        cartcontent = "Koszyk: \n"
        for product in self.shoppinglist:
            cartcontent += f"Produkt: {product.name}, Cena: {product.price}, Ilość: {product.quantity}\n"
        return cartcontent
    def __len__(self):
        sum = 0
        for product in self.shoppinglist:
            sum += product.quantity
        return sum
    def __iter__(self):
        self.index = 0
        return self

    def __next__(self):
        if self.index < len(self.shoppinglist):
            product = self.shoppinglist[self.index]
            self.index += 1
            return product
        raise StopIteration

    def add(self, name):
        for product in self.shoppinglist:
            if product.name == name:
                product.quantity += 1

    def remove(self, name):
        for product in self.shoppinglist:
            if product.name == name:
                product.quantity -= 1
                if product.quantity < 0:
                    product.quantity = 0

    def total_price(self):
        price = 0
        for product in self.shoppinglist:
            price += product.price*product.quantity
        return price

    def print_cart(self):
        for product in self.shoppinglist:
            print(product.name, product.price, product.quantity)

koszyk = Cart()
print("Dzialanie __str__ produktu: ")
print(koszyk.shoppinglist[1])
print("Dzialanie __str__ koszyka:")
print(koszyk)
print("Dzialanie __len__ koszyka: ")
print(len(koszyk))
print("\n")
print("Iteracja po koszyku: ")
for produkt in koszyk:
    print(produkt)
print("Poczatkowa wartosc koszyka:")
print(koszyk.total_price())
print("\n")
koszyk.add("marchew")
koszyk.remove("ziemniaki")
koszyk.add("jogurt")
koszyk.add("banan")
print("\n")
print("Zawartosc koszyka po wywolaniu metod:")
print(koszyk)
print("Liczba produktow po wywolaniu metod:")
print(len(koszyk))
print("Wartosc koszyka po wywolaniu metod:")
print(koszyk.total_price())
```
